```python
import requests 

import plotly.graph_objects as go
import pandas as pd

url = "http://localhost:9865/weather"
response = requests.get(
    url=url, params={
        "start_date": "2020-01-01",
        "end_date" :"2020-01-31",
        "latitude": 0,
        "longitude": 0,
    }
)
data = pd.Series(response.json())
fig = go.Figure()
fig.add_scatter(x=data.index, y= data)
```
